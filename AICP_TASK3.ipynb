```python
# Source data - dictionary to store slab information
slabs = {
    1: {"unit_range": "0 to 100", "cost_per_unit": 10, "data": [50, 150, 250]},
    2: {"unit_range": "101 to 200", "cost_per_unit": 15, "data": [100, 200, 300]},
    3: {"unit_range": "201 to 300", "cost_per_unit": 20, "data": [150, 250, 350]},
}

# Function to calculate and display cost for a given slab
def costSlab(slab_number):
    slab_info = slabs[slab_number]
    unit_range = slab_info["unit_range"]
    cost_per_unit = slab_info["cost_per_unit"]
    data = slab_info["data"]

    total_cost = sum(data) * cost_per_unit
    print(f"Bill for Slab {slab_number} (Unit Range: {unit_range}, Cost per Unit: Rs.{cost_per_unit}): Rs.{total_cost}")

# Main program
student_id = input("Enter student's ID: ")
while True:
    print(f"\nStudent ID: {student_id}")
    print("1. Display the bill of slab 1 and slab 2")
    print("2. Display the bill of slab 3")
    print("Any other key to exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        costSlab(1)
        costSlab(2)
    elif choice == '2':
        costSlab(3)
    else:
        print("Program terminated.")
        break
```
